### Es. 8.1 - 8.2
Questo esercizio ha lo scopo di andare a calcolare la funzione d'onda dello stato fondamentale di una particella libera che è soggetto ad un potenziale $$V(x)=x^4-\frac{5}{2}x^2$$
come primo approccio si utilizza un metodo variazionale Monte Carlo ovvero si parte dall' ipotizzare una famiglia di soluzioni che in questo caso è $$\Psi_T^{\sigma,\mu} = e^{-\frac{(x-\mu)^2}{2\sigma^2}}+ e^{{-\frac{(x+\mu)^2}{2\sigma^2}}}$$
e facendo variare i parametri $\sigma$ e $\mu$ si va a trovare la funzione d'onda che minimizza l'energia.
Come primo passo si implementa l'algoritmo di Metropolis che è in grado di andare a generare casualmente le posizioni $x$ secondo la distribuzione di probabilità $|\Psi_T(x)^{\sigma,\mu}|^2$, questo è utile per andare a valutare l'energia infatti vale la formula 
$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$
in cui $\frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2}$ è la distribuzione di probabilità, quindi non è altro che un integrale risolvibile con tecniche Monte Carlo.

Per andare a minimizzare l'energia si parte da parametri $\sigma$ e $\mu$ fissati e li si fa variare rispettivamente in un intervallo $[\sigma-\delta_1,\sigma+\delta_1]$ e  $[\mu-\delta_2,\mu+\delta_2]$, si vanno a calcolare le energie relative alla configurazione iniziale e a quella proposta. Se l'energia della configurazione proposta è minore allora si scelgono i nuovi parametri per ricominciare da capo altrimenti si accetta la seconda con probabilità $\exp(-\beta (E_{prop}-E_{iniz}))$ con $\beta= \frac{1}{T}$ e $T$ è una tempratura fittizia.

L'arresto dell'algoritmo avviene quando la variazione di energia è minore di un valore fissato.

Per cercare di diminuire i tempi di ricerca del minimo ho impostato inizialmente $T_i=1$, $\sigma=1$, $\mu=1$, $\delta_1=0.05$ e $\delta_2=0.1$ e abbassato la temperatura di $0.005$ fino a quando $\Delta E < 0.003$ ottenendo la seguente funzione di distribuzione caratterizzata da $\sigma = 0.630825$ e $\mu = 0.806631$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.stats as stats
```

```python
def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

# Plot a few things
plt.figure(figsize=(14,7))
scale = 0.3
plt.plot(x,(psi[0])**2,label="Solution with matrix method")
plt.xlim((-2.5,2.5))
plt.ylim((-0.2,0.6))
plt.legend()

d1=np.loadtxt("esp1/psi.dat")
plt.hist(d1,100,color="red",density=True)
plt.xlabel("x")
plt.ylabel("|psi(x)|^2")
plt.xlabel("x")
plt.grid(True)
plt.show()
```

La linea continua è la soluzione esatta trovata con il metodo delle matrici.

A questo punto ho inserito i nuovi parametri come punti di partenza ma questa volta con $\delta_1=0.005$ e $\delta_2=0.01$ e partendo da una temperatura pari a $T=0.1$ in modo tale da compiere una ricerca più precisa.

Ho ottenuto i seguenti valori: $\sigma = 0.642088$ e $\mu =  0.811247$. Di seguito sono riportati l'istogramma della distribuzione e il calcolo dell'energia tramite il metodo a blocchi con $M=10^6$ e $N=100$.

#### PIGS (T=0)

In questo esercizio si usa l'algoritmo PIGS a temperatura T=0 con la funzione costante $\psi_T(x)=1$ e la funzione d'onda che abbiamo ottenuto precedentemente. La differenza nei due approcci è visibile dai grafici delle energie:nel caso della funzione costante nelle regioni iniziali e finali del grafico dell'energia potenziale vi è un aumento mentre l'energia cinetica decresce. Questo avviene perchè la funzione costante è molto diversa dalla soluzione esatta.

L'energia totale, ottenuta invece dalla somma delle due è simile a quella del ground state ma oscilla di molto.

```python
x = np.arange(300)
En, err =np.loadtxt("potential_o.dat", usecols=(1,2), unpack=True)
En1, err1=np.loadtxt("potential_u.dat", usecols=(1,2), unpack=True)

fig=plt.figure(figsize=(14,21))
plt.subplot(311)
plt.errorbar(x, En, yerr=err, color="darkblue", label="costant psi")
plt.errorbar(x, En1, yerr=err1, color="firebrick", label="optimized psi")
plt.xlabel('M')
plt.ylabel('$<U>$')
plt.title("Potential energy ")
plt.legend()
plt.grid(True)

plt.subplot(312)
kin, errk =np.loadtxt("kinetic_o.dat", usecols=(1,2), unpack=True)
kin1, errk1=np.loadtxt("kinetic_u.dat", usecols=(1,2), unpack=True)
plt.errorbar(x, kin, yerr=errk, color="darkblue", label="costant psi")
plt.errorbar(x, kin1, yerr=errk1, color="firebrick", label="optimized psi")
plt.xlabel('M')
plt.ylabel('$<U>$')
plt.title("Kinetic energy ")
plt.legend()
plt.grid(True)

plt.subplot(313)
plt.errorbar(x, En+kin, yerr=pow(err*err+errk*errk,0.5), color="darkblue", label="costant psi")
plt.errorbar(x, En1+kin1, yerr=pow(err1*err1+errk1*errk1,0.5), color="firebrick", label="optimized psi")
plt.axhline( -0.439,0,300, color="green",label="ground state energy")
plt.xlabel('M')
plt.ylabel('$<H>$')
plt.title("Total energy ")
plt.legend()
plt.grid(True)

plt.show()
```

### PIMC 

Ora si va ad osservare il comportamento della funzioni di distribuzioni di probabilità quando si va ad aumentare la temperatura, in questo caso il sistema non ha una buona corrispondenza con il ground state poichè si trova in uno stato che è combinazione lineare dello stato fondamentale e degli altri stati eccitati e all'aumentare della temperatura lo stato del sistema diventa sempre più ortogonale allo stato fondamentale.

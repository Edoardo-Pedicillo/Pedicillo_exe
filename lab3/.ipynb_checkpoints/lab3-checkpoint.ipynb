### Esercizio 03.1

Facendo alcune ipotesi ragionevoli è possibile determinare il prezzo di un'azione $S(t)$ che può essere modellizzato da un moto Browniano geometrico ($S\sim GBM(\mu,\sigma^2)$) definito dal drift $\mu$ che indica il tasso di crescita medio del titolo e la volatilità $\sigma$, un termine stocastico che quantifica la fluttuazione del prezzo.

In questo esercizio ci si occuperà delle opzioni europee e in particolare di determinare i prezzi di call $C[S(0),0]$ e di put $P[S(0),0]$ al tempo $t=0$ attraverso un metodo Montecarlo per poi andarli a confrontare con le formule analitiche (Black-Scholes). 

I parametri di questa simulazione sono:
- prezzo dell'azione $t=0$: $S(0)=100$
- tempo di consegna: $T=1$
- prezzo concordato: $K=100$
- tasso di interesse a rischio nullo: $r=0.1$
- volatilità: $\sigma=0.25$

La stima può essere ottenuta in due modi diversi:
- usando l'espressione$$S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$$dove $W(t)\sim\mathcal{N}(0,t)$. Per calcolare il prezzo di Call si usa la formula $$C = e^{- rT } max[0 ,  S-K ]$$ ovvero il cliente compra solo se il guadagno effettivo ($max[0 ,  S-K ]$) è maggiore di quello che otterebbe se lasciasse i soldi in banca con rischio nullo. Analogamente si ha che $$P = e^{- rT } max[0 ,  K-S ]$$

- Attraverso una procedura ricorsiva ottenuta dalla formula  $$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)+\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$ con $Z_i \sim \mathcal{N}(0,1)$ si ricava il valore di $S(t)$ e si calcolano i prezzi di call e put come visto al punto precedente.

In entrambi i casi si sono stimati i prezzi e i relativi errori con il metodo a blocchi ( $10^4$ stime divise in $100$ blocchi).

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import statistics as stat 

fig=plt.figure(figsize=(14,7))

plt.subplot(221)
x, f, error = np.loadtxt("S1.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.axhline(y=14.975790778311286, color='g', linestyle='-')
plt.title('Call diretta')
plt.ylabel('prezzo')
plt.grid(True)


plt.subplot(222)
x, f, error = np.loadtxt("S2.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.axhline(y=14.975790778311286, color='g', linestyle='-')
plt.title('Call ricorsiva')
plt.ylabel('prezzo')
plt.grid(True)


plt.subplot(223)
x, f, error = np.loadtxt("P1.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.axhline(y=5.4595325819072364, color='g', linestyle='-')
plt.title('Put diretta')
plt.ylabel('prezzo')
plt.grid(True)


plt.subplot(224)
x, f, error = np.loadtxt("P2.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error)
plt.axhline(y=5.4595325819072364, color='g', linestyle='-')
plt.title('Put ricorsiva')
plt.ylabel('prezzo')
plt.grid(True)
plt.show()
```

La linea orizzontale indica il valore trovato analiticamente e si osserva un buon accordo.

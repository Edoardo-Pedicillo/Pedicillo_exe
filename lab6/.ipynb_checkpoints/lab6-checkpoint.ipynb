In questa esercitazione andiamo a simulare il modello di Ising ovvero di un sistema 1-dimensionale di $N=50$ spins che hanno un'interazione con gli elementi più vicini definita dall'Hamiltoniana 
$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$
dove $s_i$ sono gli spin che possono assumere i valori $1$ o $-1$, $J>0$ in quanto simuliamo un modello ferromagnetico. 

La simulazione viene fatta con due algoritmi diversi:
- **Algoritmo di Metropolis**: Ad ogni mossa si sceglie casualmente uno spin e si va a calcolare l'energia del sistema con e senza la sua inversione (rispettivamente $E_{old}$ ed $E_{new}$ ). Se $E_{old} > E_{new}$ allora si inverte direttamente lo spin altrimenti si accetta la mossa con probabilità $$ T({new}|{old})=exp(-\beta(E_{new}-E_{old}))$$Il tutto viene ripetuto scegliendo casualmente $50$ spins.
- **Algoritmo di Gibbs**: è simile all'algoritmo di Metropolis con la differenza che la mossa proposta viene sempre accettata. Scelto uno spin casualmente si vanno a calcolare le energie del sistema nelle due configurazioni possibili di spin e si calcola la probabilità relativa $$p=\frac{1}{1+exp(-\beta(E_{up}-E_{down}))}$$ si genera casualmente un numero $w$ compreso tra $[0,1]$, se $w<p$ si pone lo spin $+1$ altrimenti $-1$. Si ripete il tutto $50$ volte.

Con entrambi gli algoritmi si fa evolvere il sistema e si vanno a calcolare alcune grandezze macroscopiche:
- **energia interna** with $h=0$):

$$
\frac{U(N,T)}{N} = \frac{\langle H \rangle}{N}$$




- **capacità termica** (con $h=0$):

$$
C(N,T) = 
k_B\beta^2 \left(\langle H^2 \rangle -\langle H \rangle^2 \right)
$$
- **magnetizzazione** (con $h \not= 0$):
$$
M(N,T,h) =  \left\langle \sum_{i=1,N}s_i \right\rangle
$$



- **suscettibilità magnetica** (with $h=0$):

$$
\chi(N,T) = \beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle \right] 
$$


Per ciascuna di esse si va a valutare l'andamento in funzione della temperatura nell'intervallo $[0.5 , 2.0 ]$ a passi di $0.05$. Il valor medio e gli errori delle singole grandezze sono stati ottenuti suddividendo le misure in $20$ blocchi conteneti $50000$ valori e usando il metodo a blocchi. Questa parte è stata gestita con il file <font face="Courier">run.sh</font> che va a modificare la temperatura nel file  <font face="Courier">input.dat</font> e fa partire la simulazione.

Il programma è stato modificato per ripartire caricando come configurazione inizale quella contenuta nel file <font face="Courier">config.final</font> se questo è presente nella cartella, in questo modo, poichè all'inizio della simulazione si partiva da una configurazione totalmente casuale, si ha la possibilità di equilibrare il sistema con facilità facendo compiere al programma 500 step. L'equilibrazione è stata fatta solo all'inizio per portare il sistema a $T=0.5$ e non successivamente dal momento che la differenza di temperatura è molto piccola ( $0.05$ ).

In conclusione si osserva un buon accordo tra i due metodi proposti e il valore teorico atteso. Nonostante alle basse temperature il comportamento della capacità termica si discosta di poco dal valore teorico è sempre all'interno dell'intervallo di incertezza.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import statistics as stat 
```

```python
points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
```

```python
fig = plt.figure(figsize=(10,7))
plt.plot(T, e, label="Theory")
plt.title('Ising 1D, internal energy')
plt.xlabel('T')
plt.ylabel('U/N')
plt.legend()
plt.axis([0.4,2.05,-1,-0.45])

x,f,fn,error = np.loadtxt("ene.M", usecols=(0,1,2,3), delimiter=' ', unpack='true')
plt.errorbar(x,fn,yerr=error,color = "green",linestyle='dashdot', label="Metropolis")
plt.grid(True)
plt.legend()

x,f,fn,error = np.loadtxt("ene.G", usecols=(0,1,2,3), delimiter=' ', unpack='true')
plt.errorbar(x,fn,yerr=error, linestyle='dashed', label="Gibbs")
plt.grid(True)
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(9,6))
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
plt.plot(T, heat, label="Theory")
plt.title('Ising 1D, Heat capacity')
plt.xlabel('T')
plt.ylabel('C')
plt.legend()
plt.axis([0.4,2.05,0,0.5])
x,f,fn,error = np.loadtxt("Heat.M", usecols=(0,1,2,3), delimiter=' ', unpack='true')
plt.errorbar(x,fn,yerr=error,color = "g",linestyle='dashdot', label="Metropolis")
plt.grid(True)
plt.legend()

x,f,fn,error = np.loadtxt("Heat.G", usecols=(0,1,2,3), delimiter=' ', unpack='true')
plt.errorbar(x,fn,yerr=error, linestyle='dashed', label="Gibbs")
plt.grid(True)
plt.legend()
plt.show()

```

```python
fig = plt.figure(figsize=(10,7))
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
plt.plot(T, X, label="Theory")
plt.title('Ising 1D, Susceptibility')
plt.xlabel('T')
plt.ylabel('$\chi$')
plt.legend()
plt.axis([0.4,2.05,0,100])

x,f,fn,error = np.loadtxt("x.M", usecols=(0,1,2,3), delimiter=' ', unpack='true')
plt.errorbar(x,fn,yerr=error,color = "g",linestyle='dashdot', label="Metropolis")
plt.grid(True)
plt.legend()

x,f,fn,error = np.loadtxt("x.G", usecols=(0,1,2,3), delimiter=' ', unpack='true')
plt.errorbar(x,fn,yerr=error, linestyle='dashed', label="Gibbs")
plt.grid(True)
plt.legend()
plt.show()
```

```python
h=0.02 #external field
b = 1/T
fig = plt.figure(figsize=(10,7))
l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

plt.plot(T, M, label="Theory")
plt.title('Ising 1D, magnetization M with h = 0.02')
plt.xlabel('T')
plt.ylabel('$M$')
plt.legend()
plt.axis([0.4,2.05,0,0.9])

x,f,fn,error = np.loadtxt("Mag.M", usecols=(0,1,2,3), delimiter=' ', unpack='true')
plt.errorbar(x,fn,yerr=error,color = "g",linestyle='dashdot', label="Metropolis")
plt.grid(True)
plt.legend()

x,f,fn,error = np.loadtxt("Mag.G", usecols=(0,1,2,3), delimiter=' ', unpack='true')
plt.errorbar(x,fn,yerr=error, linestyle='dashed', label="Gibbs")
plt.grid(True)
plt.legend()
plt.show()
```

Dal momento che questa simulazione si basa sull'algoritmo di Metropolis vi è un problema di correlazione tra i dati che si sono ottenuti. Lo scopo di questa seconda parte è quella di capire quanto devono essere grandi i blocchi per avere dati scorrelati tra di loro. La funzione di correlazione si ottiene con la formula 
$$corr(t) = \frac{\langle v_i v_{i+t} \rangle - \langle v_i \rangle^2 }{\langle v_i^2 \rangle- \langle v_i \rangle ^2} $$

Dai grafici che seguono si osserva che un valore di $200$ elementi per blocco garantisce una non correlazione per tutti e tre gli stati.

```python
def corr(cor,v,tmin,tmax,deltat):
    k=-1
    for t in range (tmin,tmax+deltat,deltat):
        cor.append(0)
        k += 1
        sum = 0
        sum2 = 0
        prod=0
        for i in range (0,len(v)-t):
            prod += v[i]*v[i+t]
            sum += v[i]
            sum2 += v[i]*v[i]
        mean = sum/(len(v)-t)
        err2 = (sum2/(len(v)-t))-mean**2
        cor[k] = (prod/(len(v)-t)-mean**2)/err2
        
     
    return 0

epot_s = np.loadtxt("exp/solido.u", usecols=(0), skiprows=1,unpack='true') 

press_s = np.loadtxt("exp/solido.p", usecols=(0),skiprows=1,unpack='true')

epot = np.loadtxt("exp/liquido.u", usecols=(0), skiprows=1,unpack='true') 

press = np.loadtxt("exp/liquido.p", usecols=(0),skiprows=1,unpack='true')

epot_g = np.loadtxt("exp/gas.u", usecols=(0), skiprows=1,unpack='true') 

press_g = np.loadtxt("exp/gas.p", usecols=(0),skiprows=1,unpack='true')
```

```python
#########Solid
tmin =0 
tmax = 200
deltat = 1
temp = np.arange(tmin, tmax+deltat, deltat) 
corr_epot_s= arr.array('d')
j=corr(corr_epot_s,epot_s,tmin,tmax,deltat)

#autocorr press

corr_press_s= arr.array('d')
l=corr(corr_press_s,press_s,tmin,tmax,deltat)

fig=plt.figure(figsize=(14,5))

plt.subplot(121)
plt.plot(temp,corr_epot_s,)
plt.title("Autocorrelation function of epot (solid)")
plt.grid(True)


plt.subplot(122)
plt.plot(temp,corr_press_s)
plt.title("Autocorrelation function of pression (solid)")
plt.grid(True)
plt.show()
```

Ora andiamo a vedere come cambiano gli errori dell'energia potenziale e della pressione in funzione del numero di step per blocco nell'intervallo $[0, 5000 ]$. Si osserva generalmente un andamento crescente dell'errore fino a $1000$ dopodichè vi è un'oscillazione e quindi un andamento casuale, infatti a bassi step non è possibile usare il teorema del limite centrale e vi è una sottostima dell'errore, invece prendendo un numero maggiore di $1000$ le correlazioni diventano trascurabili . Per le parti successive si è fissato il valore di $2000$ step a blocco.

```python
def stat (aveN,errN,v):
    for L in x:
        M=len(v)
        N=int(M/L) 
        ave = arr.array('d')
        ave2 = arr.array('d')
        sum_prog=0
        sum_prog2=0

        for i in range (N):

            sum = 0

            for j in range(L):

                k = j+i*L
                sum += v[k]

            ave.append(sum/L)
            ave2.append((sum/L)**2)



        for i in range (N):

                sum_prog += ave[i]
                sum_prog2 += ave2 [i]


        aveN.append(sum_prog/N)
        errN.append(math.sqrt((sum_prog2/N - (sum_prog/N)**2)/N))

    return 0
```

```python
#solid

x = np.arange(10,5000,100)     
ave_Ns = arr.array('d')
err_Ns = arr.array('d')

j=stat(ave_Ns,err_Ns,epot_s)

fig=plt.figure(figsize=(14,5))

plt.subplot(121)

plt.plot(x,err_Ns)
plt.xlabel('block lenght')
plt.ylabel('error epot')
plt.title("Uncertainties of potential energy (solid)")
plt.grid(True)


aveP_Ns = arr.array('d')
errP_Ns = arr.array('d')

l=stat(aveP_Ns,errP_Ns,press_s)

plt.subplot(122)
plt.plot(x,errP_Ns)
plt.title("Uncertainties of pressure (solid)")
plt.xlabel('block lenght')
plt.ylabel('error press')
plt.grid(True)
plt.show()
```

Il passo successivo è quello di andare a calcolare la funzione di distribuzione radiale $g(r)$ che indica di quanto si discosta la distribuzione di probabilità di trovare due molecole alla distanza r rispetto ad una distribuzione uniforme alle stesse condizioni termodinamiche. Opertivamente andiamo a calcolarla nell'intervallo $[0, L/2]$ andando a suddividere questo intervallo in $100$ parti, calcolando le distanze relative di tutte le coppie di atomi e, una volta trovato il bin in cui ricade tale misura si va a sommare due al valore del bin, alla fine si normalizza dividendo per $\rho N \Delta V(r)$ dove 
$$\Delta V (r) = \frac{4\pi}{3}[(r+\delta r)^3-r^3]$$
con $\delta r$ la larghezza del bin. Anche in questo caso tramite il metodo a blocchi è possibile calcolarne il valor medio per bin e l'errore ($100$ blocchi con $2000$ step per blocco). Si può fare lo stesso calcolo anche all'interno della simulazione con l'algoritmo di dinamica molecolare NVE (200 blocchi con $10^5$ step totali) .

Si può osservare che i risultati sono molto simili nonostante si riferiscano a modelli diversi tra di loro: 
nell'algoritmo NVE il sistema si trova nel ensemble microcanonico ( energia interna fissata, ma non la temperatura) mentre in questo caso siamo nell'ensemble canonico ( a fissata temperatura). L'unica vera differenza la si trova nelle incertezze: nell' NVT sono maggiori.

L'andamento di $g(r)$ nel solido è periodica e rispecchia la periodicità del reticolo cristallino, questo effetto lo si osserva anche nei liquidi ma in maniera più rilassata mentre scompare totalmente nel gas.

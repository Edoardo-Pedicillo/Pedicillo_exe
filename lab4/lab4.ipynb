In questa sezione si va a simulare il comportamento della materia nei tre stati di aggregazione con un algoritmo di Molecular Dynamic. 

Il codice è stato opportunamente modificato per avere la possibilità di ripartire, che viene comunicata al programma attraverso una variabile booleana (restart) salvata nel file <font face="Courier">restart.dat</font>. In questo modo si da la possibilità di salvare le penultime posizioni nel file <font face="Courier">old.final</font>.

Se nella cartella è presente il file <font face="Courier">old.final</font> quando si fa partire il programma queste posizioni vengono caricate come [$\vec{r}(t-dt)$] e insieme a [$\vec{r}(t)$], che si trovano in <font face="Courier">config.final</font>, usate direttamente per calcolare le velocità e la temperatura attraverso la formula $\left\langle\frac{ mv^2}{2} \right\rangle = \frac{3}{2} kT $. Se si volesse imporre una temperatura $T^\star $bisogna calcolare il coefficiente di scalo $\alpha=\sqrt{\frac{T^\star}{T}}$, riscalare le velocità con $\alpha$ e ricalcolare le posizioni [$\vec{r}(t-dt)$].

Questo ultimo passaggio rende necessario un'equilibrazione iniziale che è stata gestita dal file  <font face="Courier">cicle.sh</font> che fa ripetere cinque volte l'intero programma andando a salvare le configurazioni finali. Come si vede dai grafici bastano $3000$ step per avere una buona equilibrazione di tutte le quantità, le successive sono state eseguite solo in via precauzionale.


Tramite l'algoritmo di Verlet è possibile simulare il comportamento delle molecole e andare a calcolare alcune quantità macroscopiche, in particolare: $\frac{U}{N},\frac{E}{N},\frac{K}{N},T$ che sono rispettivamente l'energia potenziale, l'energia totale, l'energia cinetica per particella e la temperatura. Poichè il sistema si trova in un punto di equilibrio si può simulare il campione con un numero M di passi e suddividerli in N blocchi per poter calcolarne la media e l'errore delle quantità appena menzionate tramite il metodo a blocchi.
Il numero di step nella fase di equilibrazione vale $1000$ ed è minore del nuemro di step nella fase di misurazione.

Una volta implementate queste modifiche si è provato il codice simulando il comportamento di 108 partticelle di Argon in fase solida, liquida e gassosa. I risultati sono riportati di seguito e sono espressi in SI.

### Solido

$\rho^\star = 1.1$

$T^\star = 0.8$ 

cut-off radius: $r_c = 2.2$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

epot = np.loadtxt("exp/output_epot.datS", usecols=(0),unpack='true')  
                                                                       
ekin = np.loadtxt("exp/output_ekin.datS", usecols=(0),unpack='true')

temp = np.loadtxt("exp/output_temp.datS", usecols=(0),unpack='true')

etot=np.loadtxt("exp/output_etot.datS", usecols=(0),unpack='true')
```

```python
fig=plt.figure(figsize=(14,7))
x = np.arange(5000)
plt.plot(x, epot, color="firebrick", label="Potential energy")
plt.plot(x, ekin, color="darkblue", label="kinetic energy")
plt.plot(x, temp, color="green", label="Temperature")
plt.plot(x, etot, color="yellow", label="Total Energy")

plt.xlabel('step')
plt.ylabel('Parameters')
plt.title("Equilibration")
plt.legend()
plt.grid(True)

```

```python
kB= 1.380649*10**(-23)
eps=120*kB

fig=plt.figure(figsize=(10,6))


x, f, error = np.loadtxt("exp/ave.epot.S", usecols=(0,1,2), unpack='true')
f*=eps
error*=eps
plt.errorbar(x,f,yerr=error,color="firebrick")
plt.xlabel('#step')
plt.ylabel('pot (J)')
plt.title("Potential energy")
plt.grid(True)
plt.show()

fig=plt.figure(figsize=(10,6))

x, f, error = np.loadtxt("exp/ave.ekin.S", usecols=(0,1,2), unpack='true')
f*=eps
error*=eps
plt.errorbar(x,f,yerr=error,color="darkblue")
plt.xlabel('#step')
plt.ylabel('ekin (J)')
plt.title("kinetic energy")
plt.grid(True)
plt.show()


fig=plt.figure(figsize=(10,6))

x, f, error = np.loadtxt("exp/ave.etot.S", usecols=(0,1,2), unpack='true')
f*=eps
error*=eps
plt.errorbar(x,f,yerr=error,color="yellow")
plt.xlabel('#step')
plt.ylabel('etot (J)')
plt.title("Total energy")
plt.grid(True)
plt.show()

fig=plt.figure(figsize=(10,6))

x, f, error = np.loadtxt("exp/ave.temp.S", usecols=(0,1,2), unpack='true')
f*=120
error*=120
plt.errorbar(x,f,yerr=error,color="green")
plt.xlabel('#step')
plt.ylabel('temp (K)')
plt.title("Temperature")
plt.grid(True)
plt.show()
```
